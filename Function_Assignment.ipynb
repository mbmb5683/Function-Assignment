```python
#Q1. Write a python function that takes list of numbers as input and return sum of all even numbers in the list.

def sum_of_even_numbers(*n):
  even_numbers=0
  for i in n:
    if i%2==0:
      even_numbers=even_numbers+i
  return even_numbers

sum_of_even_numbers(1,2,3,4)
```

```python
#Q2. Create a function that accepts the string and return the reverse of the string.
def string_reverse(n):
  return n[::-1]
string_reverse("hello")
```

```python
#Q3. Implement a python function that takes a list of integers and return a new list containing the squares of each number.

def square_of_numbers(n):
  new_list=[]
  for i in n:
    new_list.append(i**2)
  return new_list

square_of_numbers([1,2,3,4])
```

```python
#Q4. Write a python function that checks if the given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

is_prime(1)

```

```python
#Q5. Create an iterator class in python that generates the fibonacci sequence upto specified number of terms.

def fib(n):
  a=0
  b=1
  for i in range (n):
    yield a
    a,b=b,a+b

a=fib(10)
```

```python
#Q9. Write a python program that uses map() function to convert a list of temperature from celcius to fahrenheit.
c= [10,20,30]
cel_fahren=lambda x: x*9/5+32

list(map(cel_fahren,c))
```

```python
#Q10. Create a python program that uses filter to remove all the vowels.
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = filter(lambda char: char not in vowels, text)
    return ''.join(result)

remove_vowels("pwskills")
```
